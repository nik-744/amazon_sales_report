The following collab file contains EDA done on the amazon sales dataset.

The dataset was taken from kaggle.


```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/amazon_sale_report_clean.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['saleschannel','style','sku','asin','shippostalcode'],axis=0)
```

```python
df.head()
```

```python
df.info()
```

```python
def plot(col):
  sns.set_theme()
  sns.histplot(data=df, x=col)
  plt.xticks(rotation=90)
  plt.show()
```

```python
plot(df['category'])
```

```python
plot(df["status"])
```

```python
plot(df['size'])
```

```python
plot(df['shipstate'])
plt.figure(figsize=(15,6))
```

```python
plot(df['fulfilment'])
```
